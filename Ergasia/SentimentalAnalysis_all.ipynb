```python
voice_full_df= pd.read_csv ('/content/gdrive/MyDrive/DATA_JOUR/FINAL ESSAY/voice_full_df.csv')
efsyn_full_articles_df = pd.read_csv ('/content/gdrive/MyDrive/DATA_JOUR/FINAL ESSAY/efsyn_full.csv')
bomb_full_articles_df= pd.read_csv('/content/gdrive/MyDrive/DATA_JOUR/FINAL ESSAY/bomb_full_df.csv')
news_full_articles_df= pd.read_csv('/content/gdrive/MyDrive/DATA_JOUR/FINAL ESSAY/news_full.csv')
beast_full_df= pd.read_csv('/content/gdrive/MyDrive/DATA_JOUR/FINAL ESSAY/beast_full_df.csv')
```

```python
senti_voice_df = voice_full_df [['datetime','full_text' ]]
senti_efsyn_df = efsyn_full_articles_df [['datetime','full_text' ]]
senti_bomb_df = bomb_full_articles_df [['datetime','clean_text' ]]
senti_news_df = news_full_articles_df[['datetime','full_text' ]]
senti_newsbeast_df = beast_full_df[['datetime','full_text' ]]
```

```python
emolex_df = pd.read_csv(filepath)
emolex_df.head()
```

```python
emolex_df = emolex_df.drop_duplicates(subset=['word'])
emolex_df = emolex_df.dropna()
emolex_df.reset_index(inplace = True, drop=True)
```

```python
vec = CountVectorizer(analyzer = 'word', vocabulary = emolex_df.word,
                      lowercase=False,
                      strip_accents = 'unicode',
                      stop_words= list(nlp.Defaults.stop_words),
                      ngram_range=(1, 2))
```

```python
matrix = vec.fit_transform(senti_voice_df['full_text'])
vocab = vec.get_feature_names_out()
wordcount_df = pd.DataFrame(matrix.toarray(), columns=vocab)
wordcount_df.head()
```

```python
matrix = vec.fit_transform(senti_efsyn_df['full_text'])
vocab = vec.get_feature_names_out()
wordcount_df = pd.DataFrame(matrix.toarray(), columns=vocab)
wordcount_df.head()
```

```python
matrix = vec.fit_transform(senti_bomb_df['clean_text'])
vocab = vec.get_feature_names_out()
wordcount_df = pd.DataFrame(matrix.toarray(), columns=vocab)
wordcount_df.head()
```

```python
matrix = vec.fit_transform(senti_news_df['full_text'])
vocab = vec.get_feature_names_out()
wordcount_df = pd.DataFrame(matrix.toarray(), columns=vocab)
wordcount_df.head()
```

```python
matrix = vec.fit_transform(senti_newsbeast_df['full_text'])
vocab = vec.get_feature_names_out()
wordcount_df = pd.DataFrame(matrix.toarray(), columns=vocab)
wordcount_df.head()
```

```python
positive_words = emolex_df[emolex_df.Positive == 1]['word']
negative_words = emolex_df[emolex_df.Negative == 1]['word']
angry_words = emolex_df[emolex_df.Anger == 1]['word']
anticipation_words = emolex_df[emolex_df.Anticipation == 1]['word']
disgust_words = emolex_df[emolex_df.Disgust == 1]['word']
fear_words = emolex_df[emolex_df.Fear == 1]['word']
joy_words = emolex_df[emolex_df.Joy == 1]['word']
sadness_words = emolex_df[emolex_df.Sadness == 1]['word']
surprise_words = emolex_df[emolex_df.Surprise == 1]['word']
trust_words = emolex_df[emolex_df.Trust == 1]['word']
```

```python
senti_voice_df['anger'] = wordcount_df[angry_words].sum(axis=1)
senti_voice_df['positive'] = wordcount_df[positive_words].sum(axis=1)
senti_voice_df['joy'] = wordcount_df[joy_words].sum(axis=1)
senti_voice_df['disgust'] = wordcount_df[disgust_words].sum(axis=1)
senti_voice_df['surprise'] = wordcount_df[surprise_words].sum(axis=1)
senti_voice_df['trust'] = wordcount_df[trust_words].sum(axis=1)
senti_voice_df['anticipation'] = wordcount_df[anticipation_words].sum(axis=1)
senti_voice_df['sadness'] = wordcount_df[sadness_words].sum(axis=1)
senti_voice_df['negative'] = wordcount_df[negative_words].sum(axis=1)
senti_voice_df['fear'] = wordcount_df[fear_words].sum(axis=1)
```
