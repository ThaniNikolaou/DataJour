```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
main_url = 'https://www.news.gr'

start_url = 'https://www.news.gr/tag/syriza/page/'

start_page = 1
end_page = 2
```

```python
response = requests.get(start_url)
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
teasers_list = []

for page_num in range(start_page, end_page+1):
  page_url = start_url + str(page_num)
  response = requests.get(page_url)
  doc = BeautifulSoup(response.text, 'html.parser')

  teaser_articles_list = doc.find('div', {'class': 'main-column'}).find_all('article')


  for teaser in teaser_articles_list:
      story_dict = {}
      try:

        link = teaser.find('a')['href']


        if link.startswith("http"):
          story_dict['url'] = link
        else:
          story_dict['url'] = main_url + link
      except:
        story_dict['url'] = ''
        pass
      teasers_list.append(story_dict)
  time.sleep(1)
```

```python
news_teasers_df = pd.DataFrame(teasers_list)
news_teasers_df
```

```python
article_url = 'https://www.news.gr/politikh/article/3330237/siriza-xekinoun-i-diadikasies-gia-tin-epomeni-imera-meta-tin-paretisi-tou-alexi-tsipra.html'
```

```python
response = requests.get(article_url)
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
full_articles_list = []

for article_url in news_teasers_df['url']:
  response = requests.get(article_url)
  doc = BeautifulSoup(response.text, 'html.parser')

  article = doc.find('section', {'class': 'article__main'})

  full_article_dict = {}

  full_article_dict['site'] = "https://www.news.gr"

  full_article_dict['url'] = article_url

  try:
    section = doc.find('a', {'rel': 'category tag'}).text
    full_article_dict['section'] = section
  except:
    full_article_dict['section'] = ''
    pass

  try:
    title = doc.find('h1').text
    full_article_dict['title'] = title
  except:
    full_article_dict['title'] = ''
    pass

  try:
    date = doc.find('div', {'class': 'article-top-meta-date'}).text
    full_article_dict['date'] = date.strip()
  except:
    full_article_dict['date'] = ''
    pass


  try:
    p_texts_list = []
    paragraphs = doc.find_all('p')
    for p in paragraphs:
      p_texts_list.append(p.text)
    full_text = ' '.join(p_texts_list)
    full_text = "".join(full_text.splitlines())
    full_article_dict['full_text'] = full_text
  except:
    full_article_dict['full_text'] = ''
    pass

  full_articles_list.append(full_article_dict)
  time.sleep(1)


news_full_articles_df = pd.DataFrame(full_articles_list)
news_full_articles_df
```
