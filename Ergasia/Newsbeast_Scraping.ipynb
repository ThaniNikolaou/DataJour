```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
main_url = 'https://www.newsbeast.gr/'

start_url = 'https://www.newsbeast.gr/tag/suriza/page/'

start_page = 1
end_page = 4
```

```python
response = requests.get(start_url)
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
teasers_list = []

for page_num in range(start_page, end_page+1):
    page_url = start_url + str(page_num)
    response = requests.get(page_url)
    doc = BeautifulSoup(response.text, 'html.parser')
    teaser_articles_list = doc.find_all('div', {'class': 'feedArticleCol'})

    for teaser in teaser_articles_list:
            story_dict = {}
            try:
                link = teaser.find('a')['href']
                if link.startswith("http"):
                    story_dict['url'] = link
                else:
                    story_dict['url'] = main_url + link
            except:
                pass
            teasers_list.append(story_dict)

    time.sleep(1)
```

```python
beast_teasers_df = pd.DataFrame(teasers_list)
beast_teasers_df
```

```python
article_url = "https://www.newsbeast.gr/politiki/arthro/9837947/efi-achtsioglou-polytimos-perissotero-apo-pote-simera-o-rolos-tou-syriza"
```

```python
response = requests.get(article_url)
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
full_articles_list = []
for article_url in beast_teasers_df['url']:
  response = requests.get(article_url)
  doc = BeautifulSoup(response.text, 'html.parser')

  article = doc.find('section', {'class': 'article__main'})

  full_article_dict = {}

  full_article_dict['site'] = "https://www.newsbeast.gr/"


  full_article_dict['url'] = article_url

  try:
    section = doc.find('div', {'class': 'categories'}).text
    full_article_dict['section'] = section
  except:
    full_article_dict['section'] = ''
    pass

  try:
    title = doc.find('h1').text
    full_article_dict['title'] = title
  except:
    full_article_dict['title'] = ''
    pass

  try:
    date = doc.find('time')['datetime']
    full_article_dict['date'] = date
  except:
    full_article_dict['date'] = ''
    pass


  try:
    p_texts_list = []
    paragraphs = doc.find_all('p')
    for p in paragraphs:
      p_texts_list.append(p.text)
    full_text = ' '.join(p_texts_list)
    full_text = "".join(full_text.splitlines())
    full_article_dict['full_text'] = full_text
  except:
    full_article_dict['full_text'] = ''
    pass

  full_articles_list.append(full_article_dict)
  time.sleep(1)
```

```python
beast_full_df = pd.DataFrame(full_articles_list)
beast_full_df
```

```python
beast_full_df['datetime'] = pd.to_datetime(beast_full_df['date'], format='%Y-%m-%d %H:%M')
beast_full_df
```

```python
beast_full_df['year'] = pd.DatetimeIndex(beast_full_df['date']).year
beast_full_df['month'] = pd.DatetimeIndex(beast_full_df['date']).month
beast_full_df['day'] = pd.DatetimeIndex(beast_full_df['date']).day
beast_full_df['hour'] = pd.DatetimeIndex(beast_full_df['date']).hour
```
